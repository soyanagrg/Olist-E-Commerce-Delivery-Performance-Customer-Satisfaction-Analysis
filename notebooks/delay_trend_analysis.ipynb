```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sql_csv/monthly_late_rate.csv')

print(df.shape)
print(df.head())
```

```python
fig, ax1 = plt.subplots(figsize=(12,5))

df_plot = df[df['total_orders'] >= 1001].copy()

ax1.plot(df_plot['month'], df_plot['late_rate_pct'],
         color='#D85A30', linewidth=2, marker='o', label='Late Rate %')
ax1.set_xlabel('Month')
ax1.set_ylabel('Late Rate (%)', color='#D85A30')
ax1.tick_params(axis='x', rotation=45)

# Line 2: Review score (right axis, green)
ax2 = ax1.twinx()
ax2.plot(df_plot['month'], df_plot['avg_score'],
         color='#1D9E75', linewidth=2, marker='s', label='Avg Review Score')
ax2.set_ylabel('Avg Review Score', color='#1D9E75')

plt.title('Monthly Late Rate vs. Customer Review Score (2016-2018)')
plt.tight_layout()
plt.savefig('monthly_chart.png', dpi=150)  # saves the chart as an image file
plt.show()
```
